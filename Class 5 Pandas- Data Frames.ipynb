## Apply and Map Function

#### Apply function

```python
def fee_update(x):
    if x == 'AI':
        return 1200
    elif x == 'IOT':
        return 1500
    elif x == 'CNC':
        return 2000
    else:
        return 0
# datafeame.column-name.apply(user-defined-func)
df1.COURSE.apply(fee_update)
```

```python
df1['Updated Fee'] = df1.COURSE.apply(fee_update)
df1
```

## pd.concat
- axis > 0 (row-wise)
- axis > 1 (column-wise)

```python
df = pd.concat([df1, df2]) # by default row-wise
df
```

```python
df = pd.concat([df1, df2], axis=1) # column-wise concat
df
```

```python
df['ObtainMarks'] = df.c1 + df.c2 + df.c3 + df.c4 + df.c5 + df.c6
df
```

```python
df[['COURSE', 'c1', 'c2', 'c3', 'c4', 'c5', 'c6']]
```

```python
# AI > SUM
# IOT > PRODUCT
# CNC > CUMSUM
# OTHER > 0
def  custom_fn(x1, *l1):
    if x1 == 'AI':
        return np.sum(l1)
    elif x1 == 'IOT':
        return np.product(l1)
    elif x1 == 'CNC':
        return sum(np.cumsum(l1))
    else:
        return 0
# df['CheckValidation']
```

```python
df[['COURSE', 'c1', 'c2', 'c3', 'c4', 'c5', 'c6']].apply(lambda x:custom_fn(*x), axis=1)
```

```python
 df['CheckValidation'] = df[['COURSE', 'c1', 'c2', 'c3', 'c4', 'c5', 'c6']].apply(lambda x:custom_fn(*x), axis=1)
```
